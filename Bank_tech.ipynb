## T-Test

```python
import scipy.stats as stats
```

```python
a=df_2009_new['Patent applications, nonresidents']
b=df_2018_new['Patent applications, nonresidents']
stats.ttest_ind(a,b)
```

```python
d=df_2009_new['Patent applications, residents']
e=df_2018_new['Patent applications, residents']
stats.ttest_ind(a,b)
```

```python
cc=df_2009_new.columns
```

```python
cc
```

```python
for i in range(1,len(cc)):
  f=df_2009_new[cc[i]]
  j=df_2018_new[cc[i]]
  print(cc[i])
  statistic,pvalue=stats.ttest_ind(f,j)
  print('statistic : %s pvalue: %s' %(statistic,pvalue))
  if pvalue<0.05:
    print('significant')
  else:
    print('nonsignificant')
```
