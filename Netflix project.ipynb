Sentiment analysis of **content**

```python
dff=df[['description','release_year']]
dff_columns_name = {'description':'Description',
                    'release_year':'R_Year'
}
dff.rename(dff_columns_name,axis=1,inplace=True)
dff.head()
```

```python
from textblob import TextBlob
for index,row in dff.iterrows():
  d= row['Description']
  testmotional = TextBlob(d)
  p=testmotional.sentiment.polarity
  if p==0:
    sent='Neutral'
  elif p>0:
    sent='Positive'
  else:
    sent=='Negative'
  dff.loc[[index,2],'Sentiment'] = sent
dff=dff.groupby(['R_Year','Sentiment']).size().reset_index(name='T_Count')
```

```python
dff=dff[dff['R_Year']>2005]
DFF_Graph = px.bar(dff,x='R_Year',y='T_Count',color ='Sentiment', title='Sentiment  of content')
DFF_Graph.show()
```

Sentiment content based on listed in

```python
dff_listed=df[['listed_in','release_year']]
dff_listed_columns={
    'listed_in':'listed',
    'release_year':'R_Y'
}
dff_listed.rename(dff_listed_columns,axis=1,inplace=True)
dff_listed.columns
```

```python
for index,row in dff_listed.iterrows():
  L=row['listed']
  Testlist = TextBlob(L)
  P_L=Testlist.sentiment.polarity
  if P_L==0:
    sent = 'Netural'
  elif P_L>0:
    sent='Positive'
  else:
    sent='Negative'
  dff_listed.loc[[index,2],'Sentmen']=sent
dff_listed=dff_listed.groupby(['R_Y','Sentmen']).size().reset_index(name='ListedCount')
```

```python
dff_listed=dff_listed[dff_listed['R_Y']>2005]
List_Graph=px.bar(dff_listed,x='R_Y',y='ListedCount',color ='Sentmen',title='Sentiment  based on listed in')
List_Graph.show()
```
